```python
# Checking mail box: Seacrh key : FROM
# !pip install imapclient
import imapclient
conn=imapclient.IMAPClient('imap.gmail.com',ssl=True)
conn.login("sony.sth8@gmail.com","rbectyfhgqhnoaah")  # This is app specific password

select_info=conn.select_folder('INBOX',readonly=True)
print("Number of mails: %d " %select_info[b'EXISTS']) # Get number of mails from select information

messages=conn.search(['FROM','monica@gritfeat.com'])
print("Number of mails from Monica: %d " %len(messages))

for msgid, data in conn.fetch(messages, ['ENVELOPE']).items():
    #print(data)
    envelope = data[b'ENVELOPE']
    print('ID: %d' %msgid)
    if envelope.subject==None:
        print('Subject: ' )
    else:
        print('Subject: %s' %envelope.subject.decode())
    print('Date: %s' %envelope.date)
    print('CC: %s' %envelope.cc)
    print('BCC: %s' %envelope.bcc)
    print('Date: %s' %envelope.date)
    print('From: %s' %envelope.from_)
    print('To: {0}'.format(envelope.to))
    print("")
    #print('ID #%d: "%s" received %s' % (msgid, envelope.subject.decode(), envelope.date))
```
